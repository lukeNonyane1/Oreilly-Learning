## Analyzing Numerical Variables from the Ames Housing Datase
In this exercise, we will continue our dataset exploration by analyzing the numerical variables of this dataset. To do so, we will implement our own describe functions.

```python
import pandas as pd
```

```python
file_url = ('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter10/dataset/ames_iowa_housing.csv')
```

```python
df = pd.read_csv(file_url)
df.shape
```

```python
num_df = df.select_dtypes(include='number')
num_df.info()
```

```python
num_cols = num_df.columns
num_cols
```

```python
# create describe function
def describe_numeric(df, col_name):
    print('\nCOLUMN: {}'.format(col_name))
    print('Minimun: {}'.format(df[col_name].min()))
    print('Maxmimum: {}'.format(df[col_name].max()))
    print('Average: {}'.format(df[col_name].mean()))
    print('Standard Deviation: {}'.format(df[col_name].std()))
    print('Median: {}'.format(df[col_name].median()))
```

```python
# create for loop and call describe_numeric function for each column in num_df
for col_name in num_df:
    describe_numeric(num_df, col_name)
```

```python
describe_numeric(num_df, 'SalePrice')
```

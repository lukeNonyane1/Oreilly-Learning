## Handling Duplicates in a Breast Cancer Dataset

```python
import pandas as pd
```

```python
file_url = ('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/'\
            'master/Chapter11/dataset/breast-cancer-wisconsin.data')
```

```python
df = pd.read_csv(file_url, header=None)
```

```python
col_names = ['Sample code numner', 'Clump Thickness', 'Uniformity of Cell Size', 
            'Uniformity of Cell Shape', 'Marginal Adhesion',
            'Single Epithelial Cell Size', 'Bare Nuclei', 
            'Bland Chromatin', 'Normal Nucleoli', 'Mitoses', 'Class']
```

```python
df.columns = col_names
```

```python
df.info()
```

```python
# find number of duplicated rows
df.duplicated().sum()
```

```python
# display duplicate rows suing loc() and duplicated()
df.loc[df.duplicated()]
```

```python
df.loc[df.duplicated(keep='last')]
```

```python
# remove duplicate rows using drop_duplicates(keep='first') and assign to new df
df_unique = df.drop_duplicates(keep='first')
```

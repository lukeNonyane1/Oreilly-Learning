```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
```

```python
# instance of pre-trained model
mymodel = ResNet50()
```

```python
# load image to be classified
myimage = load_img('image.jpg', target_size=(224, 224))
```

```python
myimage
```

```python
# convert image into numpy array as the model expects
myimage = img_to_array(myimage)
```

```python
# reshape the image into four dimensional format as expected by model
myimage = myimage.reshape((1, 224, 224, 3))
```

```python
# prepare the image for submissions by running the preprocess_input()
myimage = preprocess_input(myimage)
```

```python
# run the prediction
myresult = mymodel.predict(myimage)
```

```python
# the prediction results in a number that needs to be converted into the corresponding label in text format
mylabel = decode_predictions(myresult)
```

```python
# display the label
mylabel = mylabel[0][0]
```

```python
# print the label
print('This is a : ' + mylabel[1])
```

```python
# try model with different images
myimage2 = load_img('image2.jpg', target_size=(224, 224))
myimage2
```

```python
# convert image into numpy array as the model expects
myimage2 = img_to_array(myimage2)

# reshape the image into four dimensional format as expected by model
myimage2 = myimage2.reshape((1, 224, 224, 3))

# prepare the image for submissions by running the preprocess_input()
myimage2 = preprocess_input(myimage2)

# run the prediction
myresult2 = mymodel.predict(myimage2)

# the prediction results in a number that needs to be converted into the corresponding label in text format
mylabel2 = decode_predictions(myresult2)

# display the label
mylabel2 = mylabel2[0][0]

# print the label
print('This is a : ' + mylabel2[1])
```

```python
# prepare the text to be converted into speech
sayit = "This is a "+mylabel2[1]
```

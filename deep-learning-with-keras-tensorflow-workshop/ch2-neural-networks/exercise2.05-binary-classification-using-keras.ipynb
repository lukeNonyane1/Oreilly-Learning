```python
import tensorflow as tf

import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline

# Import Keras libraries

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('data.csv')
# inspect the data
df.head()
```

```python
# visualize the data using matplotlib
plt.scatter(df[df['label'] == 0]['x1'], df[df['label'] == 0]['x2'], marker='*')
plt.scatter(df[df['label'] == 1]['x1'], df[df['label'] == 1]['x2'], marker='<')
```

```python
# prepare the data into features and labels
X_input = df[['x1', 'x2']].values
y_label = df[['label']].values
```

```python
# create a neural network model consisting of a single layer with a neuron and a sigmoig activation function
model = Sequential()
model.add(Dense(units=1, input_dim=2, activation='sigmoid'))
```

```python
# use compile method to pass additional parameters
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# use summary() to check configuration of the model before training
model.summary()
```

```python
# train model by calling fit() 
model.fit(X_input, y_label, epochs=1000)
```

```python
# use Keras's evaluate functionality to evaluate model
model.evaluate(X_input, y_label)
```
